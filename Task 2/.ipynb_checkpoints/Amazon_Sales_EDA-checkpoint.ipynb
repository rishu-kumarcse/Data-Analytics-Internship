# Amazon Sales EDA & Business Intelligence

Objective:
- Analyze Amazon sales data
- Identify trends, top products, and revenue insights
- Build business understanding using EDA

## 1. Load Dataset

```
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('amazon_sales.csv')

df.head()
df.info()
df.describe()
```

## 2. Data Cleaning

```
# Check missing values
df.isnull().sum()

# Drop or fill missing values
df = df.dropna()
```

## 3. Univariate Analysis

```
# Sales distribution
plt.hist(df['Sales'])
plt.title('Sales Distribution')
plt.show()
```

## 4. Business Questions

```
# Top 5 products by sales
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head()
print(top_products)

# Revenue by category
category_sales = df.groupby('Category')['Sales'].sum()
print(category_sales)
```

## 5. Visualization

```
# Bar chart - Top products
top_products.plot(kind='bar')
plt.title('Top Products')
plt.show()

# Category sales
category_sales.plot(kind='bar')
plt.title('Category Sales')
plt.show()
```

## 6. Insights

- Identify best-selling products
- Find highest revenue categories
- Observe sales trends
