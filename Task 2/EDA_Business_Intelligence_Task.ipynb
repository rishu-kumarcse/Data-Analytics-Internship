```python
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# 1. Clean Rating: Remove non-numeric rows and convert to float
df = df[df['rating'].apply(lambda x: x.replace('.','',1).isdigit())]
df['rating'] = df['rating'].astype(float)

# 2. Clean Category: Take only the first part of the category string
df['main_category'] = df['category'].str.split('|').str[0]

# 3. Clean Discounted Price (similar to what you did for actual_price)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
def plotting(var, num):
    plt.subplot(2, 2, num)
    sns.histplot(df[var], kde=True)
    plt.title(var)
    
    # This removes the overlapping text labels for the category axis
    if var == 'main_category' or var == 'category':
        plt.xticks([]) 

# Call the function with the cleaned columns
plt.figure(figsize=(12, 10))
plotting('main_category', 1)
plotting('actual_price', 2)
plotting('discounted_price', 3)
plotting('rating', 4)
plt.tight_layout()
plt.show()
```

```python
# Convert discount_percentage to numeric
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)

# Convert rating_count
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Fill missing values
df['rating_count'].fillna(df['rating_count'].median(), inplace=True)
```

```python
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
# =========================================
# ⚙️ 4. FEATURE ENGINEERING
# =========================================

# Discount amount
df['discount_amount'] = df['actual_price'] - df['discounted_price']

# Price bucket
df['price_bucket'] = pd.cut(df['discounted_price'],
                           bins=5,
                           labels=['Low','Medium-Low','Medium','Medium-High','High'])

df.head()
```

```python
# =========================================
# 📊 5. UNIVARIATE ANALYSIS
# =========================================

# Price distribution
plt.hist(df['discounted_price'])
plt.title('Discounted Price Distribution')
plt.show()

# Category count
sns.countplot(x='main_category', data=df)
plt.xticks(rotation=90)
plt.title('Category Distribution')
plt.show()
```

```python
# =========================================
# 🔗  MULTIVARIATE ANALYSIS
# =========================================

# Discount vs Rating
sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title('Discount vs Rating')
plt.show()

# Correlation heatmap
sns.heatmap(df[['actual_price','discounted_price','rating','rating_count']].corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# =========================================
#  BUSINESS ANALYSIS
# =========================================

# Top products
top_products = df.sort_values(by='rating_count', ascending=False).head(10)
print(top_products[['product_name','rating','rating_count']])

# Best categories
best_categories = df.groupby('main_category')['rating'].mean().sort_values(ascending=False)
print(best_categories)
```

```python
# =========================================
#  NLP - SENTIMENT ANALYSIS
# =========================================

from textblob import TextBlob

df['sentiment'] = df['review_content'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Sentiment distribution
sns.histplot(df['sentiment'])
plt.title('Sentiment Distribution')
plt.show()
```

```python
# =========================================
# 🤖 MACHINE LEARNING
# =========================================

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['actual_price','discounted_price','discount_percentage','rating_count']]
y = df['rating']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print("Model Accuracy:", model.score(X_test, y_test))
```
