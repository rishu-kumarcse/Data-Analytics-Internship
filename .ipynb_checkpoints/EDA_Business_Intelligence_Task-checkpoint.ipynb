# Exploratory Data Analysis (EDA) & Business Intelligence

Objective:
- To uncover patterns, trends, and relationships within the data.
- Develop proficiency in SQL for data extraction and dashboarding.

## 1. Descriptive Statistics & Univariate Analysis

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('your_dataset.csv')

# Basic info
df.info()

# Summary statistics
df.describe()

# Histogram
df.hist(figsize=(10,8))
plt.show()
```

## 2. SQL for Business Questions

```
# Example SQL queries (use with sqlite3 or any DB)

# Top 5 products by revenue
query1 = '''
SELECT product_name, SUM(revenue) as total_revenue
FROM sales
GROUP BY product_name
ORDER BY total_revenue DESC
LIMIT 5;
'''

# Monthly user acquisition
query2 = '''
SELECT strftime('%Y-%m', signup_date) as month, COUNT(*) as users
FROM users
GROUP BY month;
'''
```

## 3. Multivariate Analysis & Correlation

```
# Correlation heatmap
corr = df.corr()

sns.heatmap(corr, annot=True)
plt.show()

# Scatter plot
sns.scatterplot(x='feature1', y='feature2', data=df)
plt.show()
```
