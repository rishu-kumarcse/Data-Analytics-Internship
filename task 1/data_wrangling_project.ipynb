## Step 1: Load Dataset

## Step 3: Check Missing Values

```python
# Check missing values
df.isnull().sum()
```

## Step 4: Handle Missing Values

```python
# Fill numeric missing values with median
df = df.fillna(df.median(numeric_only=True))

# Fill categorical missing values with mode
for col in df.select_dtypes(include='object'):
    df[col] = df[col].fillna(df[col].mode()[0])
```

## Step 5: Remove Duplicate Rows

```python
# Remove duplicates
df = df.drop_duplicates()

df.shape
```

## Step 6: Standardize Date Columns

```python
# Example: convert purchase_date to datetime
if 'purchase_date' in df.columns:
    df['purchase_date'] = pd.to_datetime(df['purchase_date'], errors='coerce')
```

## Step 7: Feature Engineering
Example: Create Age from Date of Birth

```python
from datetime import datetime

if 'date_of_birth' in df.columns:
    df['date_of_birth'] = pd.to_datetime(df['date_of_birth'], errors='coerce')
    current_year = datetime.now().year
    df['age'] = current_year - df['date_of_birth'].dt.year
```

## Step 8: Detect Outliers (Simple Method)

```python
# Example for numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

## Step 9: Final Clean Dataset

## Step 10: Export Clean Dataset

```python
# Save cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)

print('Cleaned dataset saved successfully.')
```
